```python
industry_counts = df['Industry'].value_counts()

unique_values = df['Industry'].unique()

industry_counts = industry_counts.sort_values(ascending=False)
top_5_industries = industry_counts.head(5)

industry_counts= df[df['Industry']!='-1']['Industry'].value_counts().head(5)

industry_counts = industry_counts.sort_values(ascending=False)
top_5_industries = industry_counts.head(5)
print(top_5_industries)
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
top_5_industries.plot(kind='bar', color='skyblue')
plt.title('Top 5 Industries by Job Counts')
plt.xlabel('Industry')
plt.ylabel('Job Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Clean the "Salary Estimate" column and split into "Min Salary" and "Max Salary"
df['Salary Estimate'] = df['Salary Estimate'].str.replace('[a-zA-Z$.,\(\)]', '', regex=True)
df[['Min Salary', 'Max Salary']] = df['Salary Estimate'].str.split('-', 1, expand=True)

# Convert "Min Salary" and "Max Salary" columns to numeric
df['Min Salary'] = pd.to_numeric(df['Min Salary'], errors='coerce')
df['Max Salary'] = pd.to_numeric(df['Max Salary'], errors='coerce')

# Calculate the mean salary for each job posting
df['Mean Salary'] = (df['Min Salary'] + df['Max Salary']) / 2

# Get the top 5 industries by count
industry_counts = df['Industry'].value_counts()
industry_counts= df[df['Industry']!='-1']['Industry'].value_counts()
top_5_industries = industry_counts.sort_values(ascending=False).head(5)


# Filter the DataFrame to include only the top 5 industries
filtered_df = df[df['Industry'].isin(top_5_industries.index)]

# Calculate the mean salary for each industry
industry_mean_salary = filtered_df.groupby('Industry')['Mean Salary'].mean().reset_index()

# Create a bar plot
plt.figure(figsize=(12, 6))
plt.bar(industry_mean_salary['Industry'], industry_mean_salary['Mean Salary'])
plt.xlabel('Industry')
plt.ylabel('Mean Salary')
plt.title('Mean Salary by Industry (Top 5)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plot
plt.show()





```
